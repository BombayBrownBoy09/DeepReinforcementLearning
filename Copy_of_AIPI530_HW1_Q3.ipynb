```python
import gym
import numpy as np
from stable_baselines3 import A2C
from stable_baselines3.common.env_util import make_vec_env
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
env = gym.make('CartPole-v1')

# Use a separate environement for evaluation
eval_env = gym.make('CartPole-v1')
```

N step Advantage

```python
model = A2C('MlpPolicy', env, verbose=1, tensorboard_log="./a2c_cartpole_tensorboard/", use_n_step_advantage=True)
```

```python
# Train the agent for 70000 steps
model.learn(total_timesteps=70000)

# Evaluate the trained agent
mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=100)

print(f"mean_reward using n-step advantage:{mean_reward:.2f} +/- {std_reward:.2f}")
```

Turning off GAE

```python
model = A2C('MlpPolicy', env, verbose=1, gae_lambda=0, tensorboard_log="./a2c_cartpole_tensorboard/", use_n_step_advantage=False)
```

```python
# Train the agent for 70000 steps
model.learn(total_timesteps=70000)

# Evaluate the trained agent
mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=100)

print(f"mean_reward turning off gae:{mean_reward:.2f} +/- {std_reward:.2f}")
```

Using Default GAE

```python
# Generalized Advantage Estimator Equivalent to classic advantage when set to 1. 
model = A2C('MlpPolicy', env, verbose=1, tensorboard_log="./a2c_cartpole_tensorboard/", use_n_step_advantage=False)
```

```python
# Train the agent for 70000 steps
model.learn(total_timesteps=70000)

# Evaluate the trained agent
mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=100)

print(f"mean_reward default gae:{mean_reward:.2f} +/- {std_reward:.2f}")
```
